```python
import os
import numpy as np
from PIL import Image
import cv2
from glob import glob
from tqdm import tqdm
from natsort import natsorted
from sklearn.model_selection import train_test_split
from albumentations import HorizontalFlip, VerticalFlip, Rotate
from google.colab.patches import cv2_imshow
```

```python
TRAIN_path='/gdrive/MyDrive/TESI/TRAIN'
VALIDATION_path='/gdrive/MyDrive/TESI/VALIDATION'
```

```python
def load_train_data(train_path):
    """ Load the images and masks """
    train_x = (glob(f"{TRAIN_path}/*/images/*.png"))
    train_y = (glob(f"{TRAIN_path}/*/masks/*.png"))
 
    return train_x,train_y
```

```python
def load_val_data(val_path):
    """ Load the images and masks """
    val_x= (glob(f"{VALIDATION_path}/*/images/*.png"))
    val_y= (glob(f"{VALIDATION_path}/*/masks/*.png"))
    return val_x,val_y
```

```python
(train_x,train_y) = load_train_data(TRAIN_path)
train_x=natsorted(train_x)
train_y=natsorted(train_y)
print(len(train_x))
print(len(train_y))
```

```python
(val_x,val_y) = load_val_data(VALIDATION_path)
val_x=natsorted(val_x)
val_y=natsorted(val_y)
print(len(val_x))
print(len(val_y))
```

```python
H = 512
W = 512
for idx, (x, y) in tqdm(enumerate(zip(train_x, train_y)), total=len(train_x)):
        """ Extracting the dir name and image name """
        dir_name = x.split("/")[-3]
        name = dir_name + "_" + x.split("/")[-1].split(".")[0]

        """ Read the image and mask """
        x = cv2.imread(x, cv2.IMREAD_GRAYSCALE)
        y = cv2.imread(y, cv2.IMREAD_GRAYSCALE)
        #cv2_imshow(y)
        #cv2.waitKey(0)
        X = [x]
        Y = [y]
        idx = 0
        for i, m in zip(X, Y):
            i = cv2.resize(i, (W, H),interpolation = cv2.INTER_CUBIC)
            m = cv2.resize(m, (W, H),interpolation = cv2.INTER_CUBIC)
            m=m/255.0
            m=(m>0.5)*255
  
            #cv2_imshow(i)
            #cv2.waitKey(0)
            #cv2.destroyAllWindows()
            #i = cv2.convertScaleAbs(i, alpha=(255.0))
            tmp_image_name = f"{name}.png"
            tmp_mask_name  = f"{name}.png"

            image_path = os.path.join('/gdrive/MyDrive/TESI/output/train/train_images/images/', tmp_image_name)
            mask_path  = os.path.join('/gdrive/MyDrive/TESI/output/train/train_masks/masks/', tmp_mask_name)
            print(tmp_image_name)

            cv2.imwrite(image_path,i, [int(cv2.IMWRITE_PNG_COMPRESSION),0])
            cv2.imwrite(mask_path, m, [int(cv2.IMWRITE_PNG_COMPRESSION),0])

            idx += 1
```
